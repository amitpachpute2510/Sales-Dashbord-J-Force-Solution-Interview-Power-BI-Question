```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

sales = pd.read_sql("SELECT * FROM sales", conn)
customers = pd.read_sql("SELECT * FROM customers", conn)
products = pd.read_sql("SELECT * FROM products", conn)

print(sales.shape)
print(customers.shape)
print(products.shape)
```

```python
print(sales.columns)
print(customers.columns)
print(products.columns)
```

```python
F_DF = (
    sales.merge(customers,on='customer_id',how='left')
    .merge(products,on ='product_id',how='left')   
)
print(F_DF.head())
```

```python
# Create a new column in the sales data to calculate total_sale_amount (quantity sold × unit price).
F_DF['TOTAL_SALES'] =( F_DF['quantity_sold']*F_DF['price'])
print(F_DF.columns)

print("\nTOTAL_SALES Column Added")
```

```python
# CLEANING
# IN AGE COLUMN THERE ARE NEGATIVE AGE ARE SO THAT CONVERTED TO NULL

print(F_DF['age'])
```

```python
# CLEANING
# IN AGE COLUMN THERE ARE NEGATIVE AGE ARE SO THAT CONVERTED TO NULL
F_DF.loc[F_DF['age']<0,'age']=None
```

```python
print(F_DF['age'])
```

```python
# data type of date Column is Diffrent 
# SO ,

F_DF['sale_date'] = pd.to_datetime(F_DF['sale_date'])
F_DF['join_date'] = pd.to_datetime(F_DF['join_date'])
```

```python
print(F_DF['sale_date'])
print(F_DF['join_date'])
```

```python
# Remove Duplicates
F_DF.drop_duplicates()
```
